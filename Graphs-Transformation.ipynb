```python
import numpy as np

import matplotlib.pyplot as plt
import random 
import copy 
```

```python
sample = 1000
Fs = sample

DPI_VAL = None

ORIGINAL_COLOR = 'b'
TRANSFORMED_COLOR = 'r'
ZERO_COLOR = 'k'


ORIGINAL_WAVE_LABEL = "Original Wave"
TRANSFORMED_WAVE_LABEL = "Transformed Wave"


x = np.arange(sample)
f = x/100
y = np.sin(2 * np.pi * f * x / Fs)/2

def no_ticks():
    plt.tick_params(
        axis='x',         
        which='both',     
        bottom=False,     
        labelbottom=False)

    plt.tick_params(
        axis='y',        
        which='both',    
        left=False, 
        labelleft=False) 

LIMIT = 3
no_ticks()
plt.ylim(top=1 * LIMIT, bottom=-1 * LIMIT)
plt.plot(x, y, color=ORIGINAL_COLOR)
plt.axhline(y=0, color=ZERO_COLOR, linestyle='--')

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("Original Sound Wave")
# plt.legend()
plt.tight_layout()
# plt.savefig('original.png', format='png', dpi=DPI_VAL, bbox='tight')
plt.show()
```

```python
# Uniform noise addition
NOISE_RANGE = 0.1
no_ticks()
SHIFT_GRAPH = LIMIT/2
plt.ylim(top=1 * LIMIT, bottom=-1 * LIMIT)
y_transform = copy.deepcopy(y)
y_transform = [elem + np.random.normal(loc=0, scale=NOISE_RANGE) for elem in y_transform]# add noise
# y_transform = [elem + random.uniform(NOISE_RANGE*-1, NOISE_RANGE) for elem in y_transform]# add noise

y_transform = np.array(y_transform)


plt.axhline(y=SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=-SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=0, linestyle='-', color=ZERO_COLOR)

plt.plot(x, y + SHIFT_GRAPH, color=ORIGINAL_COLOR, label=ORIGINAL_WAVE_LABEL)
plt.plot(x, y_transform - SHIFT_GRAPH, color=TRANSFORMED_COLOR, label=TRANSFORMED_WAVE_LABEL)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("Uniform noise addition")
plt.legend()
plt.tight_layout()
# plt.savefig('uniform.png', format='png', dpi=DPI_VAL, bbox='tight')
plt.show()
```

```python
# Drop/Frame
y_transform = copy.deepcopy(y)
SHIFT_GRAPH = LIMIT/2
plt.ylim(top=1 * LIMIT, bottom=-1 * LIMIT)
no_ticks()
NUM_CLIPS = 5
START_VAL = 20
CLIP_RANGE = 40
for i in range(NUM_CLIPS):
    min_val = sample/NUM_CLIPS * i + START_VAL
    max_val = min_val + CLIP_RANGE
    for i in range(int(min_val), int(max_val)):
        y_transform[i] = 0
        
plt.axhline(y=SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=-SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=0, linestyle='-', color=ZERO_COLOR)
# plt.plot(x, y- SHIFT_GRAPH, color=TRANSFORMED_COLOR, linestyle="--")

plt.plot(x, y + SHIFT_GRAPH, color=ORIGINAL_COLOR, label=ORIGINAL_WAVE_LABEL)
plt.plot(x, y_transform - SHIFT_GRAPH, color=TRANSFORMED_COLOR, label=TRANSFORMED_WAVE_LABEL)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("Frame Drop")
plt.legend()
plt.tight_layout()
# plt.savefig('dropframe.png', format='png', dpi=DPI_VAL, bbox='tight')
plt.show()

```

```python
#Scaling
SCALING_FACTOR = 0.8
y_transform = 0.5*np.sin(2 * np.pi * SCALING_FACTOR * f * x / Fs)
SHIFT_GRAPH = LIMIT/2

plt.ylim(top=1 * LIMIT, bottom=-1 * LIMIT)
no_ticks()

x_transform = np.arange(sample/SCALING_FACTOR, step=1/SCALING_FACTOR)
# print(x_transform)


plt.axhline(y=SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=-SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=0, linestyle='-', color=ZERO_COLOR)

# plt.plot(x, y + SHIFT_GRAPH , color=ORIGINAL_COLOR)
# plt.plot(x_transform, y_transform - SHIFT_GRAPH, color=TRANSFORMED_COLOR)

plt.plot(x, y + SHIFT_GRAPH, color=ORIGINAL_COLOR, label=ORIGINAL_WAVE_LABEL)
plt.plot(x_transform, y_transform - SHIFT_GRAPH, color=TRANSFORMED_COLOR, label=TRANSFORMED_WAVE_LABEL)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("Frequency Scaling")
plt.legend()
plt.tight_layout()
# plt.savefig('frequency.png', format='png', dpi=DPI_VAL, bbox='tight')
plt.show()



```

```python
# Amplitude Modification
AMP_SCALE_FACTOR = 0.5
y_transform = AMP_SCALE_FACTOR*copy.deepcopy(y)



SHIFT_GRAPH = LIMIT/2
plt.ylim(top=1 * LIMIT, bottom=-1 * LIMIT)
no_ticks()

plt.axhline(y=SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=-SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=0, linestyle='-', color=ZERO_COLOR)

plt.plot(x, y + SHIFT_GRAPH, color=ORIGINAL_COLOR, label=ORIGINAL_WAVE_LABEL)
plt.plot(x, y_transform - SHIFT_GRAPH, color=TRANSFORMED_COLOR, label=TRANSFORMED_WAVE_LABEL)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("Amplitude Modification")
plt.legend()
plt.tight_layout()
# plt.savefig('amplitude.png', format='png', dpi=DPI_VAL, bbox='tight')
plt.show()

```

```python
# #Clipping
CIPPING_VAL = 0.25
y_transform = [min(CIPPING_VAL, elem) for elem in y]
y_transform = [max(-1*CIPPING_VAL, elem) for elem in y_transform]
y_transform = np.array(y_transform)

SHIFT_GRAPH = LIMIT/2
plt.ylim(top=1 * LIMIT, bottom=-1 * LIMIT)
no_ticks()

plt.axhline(y=SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=-SHIFT_GRAPH, color=ZERO_COLOR, linestyle='--')
plt.axhline(y=0, linestyle='-', color=ZERO_COLOR)

# plt.plot(x, y - SHIFT_GRAPH, color=TRANSFORMED_COLOR, linestyle='--')
plt.axhline(y=-SHIFT_GRAPH+CIPPING_VAL, color=TRANSFORMED_COLOR, linestyle='--')
plt.axhline(y=-SHIFT_GRAPH-CIPPING_VAL, color=TRANSFORMED_COLOR, linestyle='--')


plt.plot(x, y + SHIFT_GRAPH, color=ORIGINAL_COLOR, label=ORIGINAL_WAVE_LABEL)
plt.plot(x, y_transform - SHIFT_GRAPH, color=TRANSFORMED_COLOR, label=TRANSFORMED_WAVE_LABEL)

# plt.plot(x, y_transform - SHIFT_GRAPH, color=TRANSFORMED_COLOR)

# plt.plot(x, y + SHIFT_GRAPH , color=ORIGINAL_COLOR)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("Clipping")
plt.legend()
plt.tight_layout()
# plt.savefig('clipping.png', format='png', dpi=DPI_VAL, bbox='tight')
plt.show()




# #High pass
# y_transform = y
# for i in range(0,400):
#     y_transform[i] = 0

#Low pass
# y_transform = y
# for i in range(700,1000):
#     y_transform[i] = 0

```
